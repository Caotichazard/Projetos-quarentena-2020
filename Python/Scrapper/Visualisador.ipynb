```python
import ploter
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dt
```

```python
nomes = ploter.read_from_meta()
```

```python
print(nomes)
```

```python
precos,datas,nome_prod,tracking = ploter.read_from_prod(nomes[1])
if tracking:
    plt.figure(1, figsize=(10,4))
    plt.plot(datas,precos)
    plt.gcf().autofmt_xdate()
    plt.title(nome_prod)
    plt.show()
```

```python
precos,datas,nome_prod,tracking = ploter.read_from_prod(nomes[2])
plt.figure(1, figsize=(10,4))
plt.plot(datas,precos)
plt.gcf().autofmt_xdate()
plt.title(nome_prod)
plt.show()
```

```python
for nome in nomes:
    precos,datas,nome_prod,tracking = ploter.read_from_prod(nome)
    if tracking:
        plt.plot(datas,precos)
        plt.gcf().autofmt_xdate()
        plt.title(nome_prod)
        plt.show()
    
```
